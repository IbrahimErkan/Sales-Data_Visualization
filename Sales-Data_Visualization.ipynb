```python
x = range(1,44)
y = veri["Unit Price"]

plt.figure(figsize = (15,8))
plt.xticks(np.arange(1,44,1))
plt.bar(x,y, color = "red", width = 0.5)
plt.show()
```

```python
x = range(1,44)
y = veri["Unit Price"]

plt.figure(figsize = (15,8))
plt.bar(x,y, color = "red", width = 0.5)
plt.xlabel("Günler")
plt.ylabel("Birim Fiyat")
plt.title("Günlük Değişen Birim Fiyat")
plt.xticks(np.arange(1,44,1))
plt.plot(x,y, color = "black")
plt.show()
```
